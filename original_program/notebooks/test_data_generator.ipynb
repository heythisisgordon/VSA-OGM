```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pickle as pkl
```

```python
np.random.seed(0)

EMPTY_NOISE = 0.00
LIDAR_RANGE = 4
OCCUPIED_NOISE = 0.05
NUM_POINTS = 5000
WORLD_SIZE = [-6, 6, -6, 6]
```

```python
# ---------------------------------------------
# Explore how to draw are circle in with NumPy
# ---------------------------------------------
thetas = np.random.rand(NUM_POINTS) * 2 * np.pi

occ_lidar_x: np.ndarray = (LIDAR_RANGE) * np.cos(thetas)
occ_lidar_y: np.ndarray = (LIDAR_RANGE) * np.sin(thetas)

occ_lidar_x += np.random.normal(0, OCCUPIED_NOISE, (NUM_POINTS))
occ_lidar_y += np.random.normal(0, OCCUPIED_NOISE, (NUM_POINTS))

plt.plot(occ_lidar_x, occ_lidar_y, "yo", markersize=1)
plt.axis("equal")
plt.xlim(WORLD_SIZE[0], WORLD_SIZE[1])
plt.ylim(WORLD_SIZE[2], WORLD_SIZE[3])
plt.show()
```

```python
# ---------------------------------------------
# Explore how to draw are circle in with NumPy
# ---------------------------------------------
thetas = np.random.rand(NUM_POINTS) * 2 * np.pi

lidar_ranges = np.sqrt(np.random.rand(NUM_POINTS)) * (LIDAR_RANGE - 0.09)

empty_lidar_x: np.ndarray = lidar_ranges * np.cos(thetas)
empty_lidar_y: np.ndarray = lidar_ranges * np.sin(thetas)

plt.plot(empty_lidar_x, empty_lidar_y, "bo", markersize=1)
plt.axis("equal")
plt.xlim(WORLD_SIZE[0], WORLD_SIZE[1])
plt.ylim(WORLD_SIZE[2], WORLD_SIZE[3])
plt.show()
```

```python
plt.plot(occ_lidar_x, occ_lidar_y, "yo", markersize=1)
plt.plot(empty_lidar_x, empty_lidar_y, "bo", markersize=1)
plt.axis("equal")
plt.xlim(WORLD_SIZE[0], WORLD_SIZE[1])
plt.ylim(WORLD_SIZE[2], WORLD_SIZE[3])
plt.show()
```

```python
data_dict = {
    "occ_lidar_x": occ_lidar_x,
    "occ_lidar_y": occ_lidar_y,
    "empty_lidar_x": empty_lidar_x,
    "empty_lidar_y": empty_lidar_y,
}

data_sp: str = os.path.join(".", "ablation_data.pkl")
with open(data_sp, "wb") as f:
    pkl.dump(data_dict, f)
```
